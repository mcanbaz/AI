### NLP - Pre-processing

```python
df['text'].replace({'r"\&\#[0-9]+\;"':''},regex=True,inplace=True)
# remove HTML chars
```

```python
df['text']=df['text'].str.lower()
# convert lower case
```

```python
df['text'].replace({'[^\w\s]':''},regex=True,inplace=True)
# remove punctuation
```

```python
df['text'].replace({'\d+':''},regex=True,inplace=True)
# remove the numbers
```

```python
df['text'].replace({'\n':'','\r':'',"\'":""},regex=True,inplace=True)
# Line break marks are removed
```

```python
df['text'].replace({'r"[\s]+"':''},regex=True,inplace=True)
# remove whitespaces
```

```python
from unicodedata import normalize
df['text']=df['text'].apply(lambda text: normalize("NFKD", str(text)).encode("ascii", "ignore").decode("utf-8", "ignore"))
# accent is removed
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
df['text']=df['text'].apply(lambda x:" ".join([i for i in str(x).split(" ")  if i not in stop_words]))
# Removed unnecessary words
```

```python
# Language Detecting
from langdetect import detect
df['language']=df['text'].apply(detect)
df=df[df['language']=='en']
df
```

```python
# Polarity - Subjectivity
from textblob import TextBlob
df[['polarity','subjectivity']]=df['text'].apply(lambda t:pd.Series(TextBlob(t).sentiment))
```

```python
# Tokenization
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
df['tokenized']=[tokenizer.tokenize(i) for i in df['text'].values]
```

```python
#import nltk
#nltk.download('wordnet')
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
df['lemmatize']=[[lemmatizer.lemmatize(str(i)) for i in words] for words in df['text'].values]
```

```python
df.loc[df['subjectivity']>0.5,'sentiment']=2 #positive
df.loc[df['subjectivity']==0.5,'sentiment']=1 #neutral
df.loc[df['subjectivity']<0.5,'sentiment']=0 #negative
df
```

```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
stopwords = set(STOPWORDS)

def generateWordcloud(text,tag):
    my_mask = np.array(Image.open('img/cloud.png'))    
    wc = WordCloud(background_color='white',
              mask=my_mask,
              collocations=False,
              width=800,
              height=600,
              max_words=1000,
              #contour_color='blue',
              #contour_width=1,
              stopwords=stopwords)
    
    wc.generate(str(text))
    wc.to_file("img/cloud-"+tag+".png")
    # print("Word Cloud Saved Successfully")
    path="img/cloud-"+tag+".png"
    display(Image.open(path))
```

```python
generateWordcloud(df[df['sentiment']==1].text,'neutral')
```

```python
generateWordcloud(df[df['sentiment']==2].text,'positive')
```

```python
generateWordcloud(df[df['sentiment']==0].text,'negative')
```
