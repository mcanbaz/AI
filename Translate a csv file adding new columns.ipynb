```python
import pandas as pd
import numpy as np

from googletrans import Translator
translator = Translator()
translations = {}


df = pd.read_csv("articles.csv", encoding='latin1')
df.head()
```

```python
df['Title_tr'] = df['Title'].apply(translator.translate, src='en', dest='tr').apply(getattr, args=('text',))
df['Article_tr'] = df['Article'].apply(translator.translate, src='en', dest='tr').apply(getattr, args=('text',))
df
```

```python
df.to_csv(r'article_tr.csv', index = False)
```
